# Keypoint Extraction — Run on Kaggle

This script extracts YOLOv8-pose keypoints from all UCF-Crime frames and saves them to a CSV file.

**Run the notebook directly on Kaggle (free GPU):**
https://www.kaggle.com/code/faizzzan/surveillance

---

## Why Kaggle?

The full dataset has 670k+ frames. Running locally on CPU takes 10+ hours.
Kaggle gives you a free T4 GPU which cuts it down to 2-3 hours.

---

## How to Run

1. Create a free account at kaggle.com
2. Go to Settings → Phone Verification → verify your number (unlocks 30hr/week free GPU)
3. Open the notebook: https://www.kaggle.com/code/faizzzan/surveillance
4. Click Edit in the top right
5. Add Input → Datasets → search `ucf-crime-dataset` by Sanskar Hasija → add it
6. Settings → Accelerator → select GPU T4 x2
7. Settings → Internet → On
8. Click Run All
9. When done, download `keypoints.csv` from the Output panel
10. Place it in `data/processed/keypoints.csv` in your local project

---

## Output

A CSV file with one row per frame containing:

- `frame_path` — path to original frame
- `label` — activity class (Fighting, Normal, Robbery etc.)
- `split` — Train or Test
- `kp_0` to `kp_33` — 34 keypoint coordinates extracted by YOLOv8-pose

---

## Notes

- `FRAME_SKIP = 3` means every 3rd frame is processed — reduces compute by 3x
- Only the most confident person per frame is kept
- Frames where no person is detected are skipped
